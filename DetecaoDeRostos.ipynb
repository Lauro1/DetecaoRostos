```python
import cv2
from google.colab.patches import cv2_imshow

# Caminho da imagem carregada
image_path = "formatura2001.png"  # Nome do arquivo carregado
image = cv2.imread(image_path)   # Lê a imagem em formato colorido (BGR)

# Mostrar a imagem colorida
cv2_imshow(image)
```

```python
# Carregar o classificador Haar para detecção de rostos
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Converter para escala de cinza para a detecção
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detectar rostos
faces = face_cascade.detectMultiScale(
    gray_image,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30)
)
print(f"Rostos detectados: {len(faces)}")

# Desenhar retângulos ao redor dos rostos detectados
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)  # Azul para os retângulos

# Mostrar a imagem com os rostos destacados
cv2_imshow(image)
```

```python
# Salvar a imagem processada com os rostos destacados
cv2.imwrite('detected_faces_formatura2001.png', image)
```
